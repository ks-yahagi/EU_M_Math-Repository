```python
import numpy as np
import numpy.random as random
import matplotlib as mp1
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import math

plt.figure(figsize=(20,6))

#random.seed(0)
#plt.subplot(2,1,1)
#plt.hist(np.random.uniform(0.0,1.0,10),bins=60,range(0,1))

#解答を見た
def uni_hist(N):
  #一様乱数の発生
  x= np.random.uniform(0.0,1.0,N)
  y= np.random.uniform(0.0,1.0,N)

  plt.subplot(2,1,1)
  plt.hist(x)
  plt.title("No1:histogram")

  plt.subplot(2,1,2)
  plt.hist(y)
  plt.title("No2:histogram")
  
  plt.grid(True)

  #タイトルの被りを防ぐ
  plt.tight_layout()

uni_hist(1000)
```
