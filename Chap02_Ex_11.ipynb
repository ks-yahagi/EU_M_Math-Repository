```python
import numpy as np
import matplotlib as mp1
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(20,6))

x=np.linspace(-10,10,100)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

plt.title('sin and cos')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```
